```python
import pandas as pd
from scipy.io.arff import loadarff
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
```

```python
def Gaussian_Distribution(N=2, M=1000, m=0, sigma=1):
    '''
    Parameters
    ----------
         N dimension
         M number of samples
         m sample mean
         sigma: sample variance
    
    Returns
    -------
         data shape(M, N), M samples of N-dimensional Gaussian distribution
         Gaussian Gaussian distribution probability density function
    '''
    mean = np.zeros(N) + m  # Mean matrix, the mean of each dimension is m
    cov = np.eye(N) * sigma  # Covariance matrix, the variance of each dimension is sigma

    # Generate N-dimensional Gaussian distribution data
    data = np.random.multivariate_normal(mean, cov, M)
    # N-dimensional data Gaussian distribution probability density function
    Gaussian = multivariate_normal(mean=mean, cov=cov)
    
    return data, Gaussian
```

```python
M = 100
data, Gaussian = Gaussian_Distribution(N=2, M=M, sigma=0.1)
# Generate a two-dimensional grid plane
X, Y = np.meshgrid(np.linspace(-1,1,M), np.linspace(-1,1,M))
# Two-dimensional coordinate data
d = np.dstack([X,Y])
# Calculate the two-dimensional joint Gaussian probability
Z = Gaussian.pdf(d).reshape(M,M)


'''Binary Gaussian probability distribution map'''
fig = plt.figure(figsize=(6,4))
ax = plt.axes(projection = '3d')
ax.scatter(X, Y, Z)
# ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='seismic', alpha=0.8)
# ax.set_xlabel('X')
# ax.set_ylabel('Y')
# ax.set_zlabel('Z')
plt.show()
```

```python
def distance(p, q, matrix):
    sm = 0
    for key in range(matrix.shape[-1]):
        dsq = (matrix[p][key] - matrix[q][key]) ** 2
        sm += dsq
    return np.sqrt(sm)
```

```python
def baseline(dim):
    baseline_model = nn.Sequential(
        nn.Linear(dim, 256, dtype=torch.float64),
        nn.ReLU(),
        nn.Linear(256, 64, dtype=torch.float64),
        nn.ReLU(),
        nn.Linear(64, 64, dtype=torch.float64),
        nn.ReLU(),
        nn.Linear(64, 256, dtype=torch.float64),
        nn.ReLU(),
        nn.Linear(256, 1, dtype=torch.float64),
        nn.Sigmoid()
    )

    return baseline_model
```

```python
def find_manifold_loss(f_data, f_x):
    N = len(f_data)
    graph = np.zeros((N, N))
    # print(graph.shape)
    for i in range(N):
        for j in range(i, N):
            val = distance(i, j, f_data.numpy())
            graph[i][j] = val
            graph[j][i] = val

    # print(graph.shape)

    graph_new = graph/graph.max()
    # graph_new = graph
    adjacency_matrix = np.ones((N, N))
    for i in range(N):
        adjacency_matrix[i][i] = 0
    # adjacency_matrix = np.zeros((N, N))
    # for i in range(N):
    #     adjacency_matrix[i][i] = N
    laplasian = adjacency_matrix - graph_new

    part_1 = np.dot(f_x.detach().numpy().T, laplasian)
    loss = np.dot(part_1, f_x.detach().numpy())

    return loss.reshape(-1)[0]
    # return loss.reshape(-1)[0], laplasian, graph_new
```
